# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
# Transform the data
totals = data.groupby('ItemType')[['RetailSales', 'RetailTransfers', 'WarehouseSales']].sum()

# Plot
totals.plot(kind='bar', figsize=(14,8))
plt.title('Total Retail Sales, Retail Transfers and Warehouse Sales by Item Type', fontweight=700)
plt.ylabel('Total')
plt.show()
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
supplier_sales = data.groupby('Supplier')[['RetailSales', 'RetailTransfers', 'WarehouseSales']].sum()
supplier_sales['Total Sales'] = supplier_sales.sum(axis=1)
#top 10 suppliers with the most total sales
top_10_suppliers = supplier_sales.nlargest(10, 'Total Sales')

top_10_suppliers['Retail Sales Mix'] = (top_10_suppliers['RetailSales'] / top_10_suppliers['Total Sales']) * 100
top_10_suppliers['Retail Transfers Mix'] = (top_10_suppliers['RetailTransfers'] / top_10_suppliers['Total Sales']) * 100
top_10_suppliers['Warehouse Sales Mix'] = (top_10_suppliers['WarehouseSales'] / top_10_suppliers['Total Sales']) * 100

plt.figure(figsize=(10, 6))
sns.barplot(x='Total Sales', y=top_10_suppliers.index, data=top_10_suppliers, orient='h')
plt.xlabel('Total Sales')
plt.ylabel('Supplier')
plt.title('Sales Mix for Top 10 Suppliers with Most Total Sales')
plt.show()
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
monthly_avg_sales = data.groupby(['Year', 'Month'])[['RetailSales', 'RetailTransfers', 'WarehouseSales']].mean()
monthly_avg_sales.reset_index(inplace=True)

plt.figure(figsize=(10, 6))
for column in ['RetailSales', 'RetailTransfers', 'WarehouseSales']:
    plt.plot(monthly_avg_sales.index, monthly_avg_sales[column], label=column)

# Adding labels and title
plt.xlabel('Year-Month')
plt.ylabel('Average Sales')
plt.title('Average Sales per Month')
plt.show()
```

## 4. Plot the same information as above but as a bar chart.

```python
monthly_avg_sales = data.groupby(['Year', 'Month'])[['RetailSales', 'RetailTransfers', 'WarehouseSales']].mean()
monthly_avg_sales.reset_index(inplace=True)

plt.figure(figsize=(12, 6))

# Plotting 
plt.bar(monthly_avg_sales.index, monthly_avg_sales['RetailSales'], width=0.2, label='Retail Sales')
plt.bar(monthly_avg_sales.index, monthly_avg_sales['RetailTransfers'], width=0.2, label='Retail Transfers')
plt.bar(monthly_avg_sales.index, monthly_avg_sales['WarehouseSales'], width=0.2, label='Warehouse Sales')


plt.xlabel('Year-Month')
plt.ylabel('Average Sales')
plt.title('Average Retail Sales, Retail Transfers, and Warehouse Sales per Month')
plt.xticks(monthly_avg_sales.index, monthly_avg_sales.index)
plt.show()
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
grouped_data = data.groupby(['Year', 'Month', 'ItemType'])['RetailSales'].sum().reset_index()

# Creating a multi-line chart with a line for each Item Type
plt.figure(figsize=(20, 12))

# Plotting a line for each Item Type
grouped_data.pivot_table(index=['Year', 'Month'], columns='ItemType', values='RetailSales').plot()


# Adding labels and title
plt.xlabel('Year-Month')
plt.ylabel('Retail Sales')
plt.title('Retail Sales by Item Type over Time')
plt.legend(loc='upper left')
plt.show()
```

## 6. Plot the same information as above but as a bar chart.

```python
grouped_data = data.groupby(['Year', 'Month', 'ItemType'])['RetailSales'].sum().reset_index()

# Creating a multi-line chart with a line for each Item Type
plt.figure(figsize=(12, 6))

# Plotting a line for each Item Type
for item_type, item_type_data in grouped_data.groupby('ItemType'):
    plt.bar(item_type_data['Year'].astype(str) + '-' + item_type_data['Month'].astype(str),
             item_type_data['RetailSales'], label=item_type)

# Adding labels and title
plt.xlabel('Year-Month')
plt.ylabel('Retail Sales')
plt.title('Retail Sales by Item Type over Time')
plt.legend(loc='upper left')
plt.show()
```
