```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

**Hint**: Check out the nrows, ncols and index arguments of subplots. 

Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# Method 1: Using subplot method
fig, ax = plt.subplots(figsize=(15,6))
ax.set(xlabel="x", ylabel="y", title="Line Chart")
plt.plot(x,y, color='blue', linewidth=2, linestyle='-', label='y = 2x')
plt.show()
ax.set(xlabel="x", ylabel="z", title="Quadratic Chart")
plt.plot(x,z, color='green', linewidth=2, linestyle='--', label='z = x^2')
plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
#Approach 2: Using subplots with nrows and ncols arguments

fig,(ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=False, figsize=(15,5))

ax0.set_title(" Line Chart")
ax0.set_xlabel("x")
ax0.set_ylabel("y")
ax0.plot(x, y, color='pink', linewidth=2, linestyle='-', label='y = 2x')

ax1.set_title(" Quadratic Chart")
ax1.set_xlabel("x")
ax1.set_ylabel("z")
ax1.plot(x, z, color='blue', linewidth=2, linestyle='--', label='z = x^2')

plt.show()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 100)
y_squared = x ** 2
y_exp = np.exp(x)

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Plot y = x^2 in the first subplot (normal scale)
axes[0].plot(x, y_squared, color='blue', label='y = x^2')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('y = x^2 (Normal Scale)')

# Plot y = e^x in the second subplot (logarithmic scale)
axes[1].plot(x, y_exp, color='green', label='y = e^x')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].set_title('y = e^x (Logarithmic Scale)')

# Set logarithmic scale for the y-axis in the second subplot
axes[1].set_yscale('log')

# Display the plots
plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
fitbit.groupby('Months_encoded')['Steps'].median().plot(kind='bar', color='skyblue')
plt.xlabel('Month')
plt.ylabel('Median Steps')
plt.title('Steps by Month')
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])


# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
plt.subplot(1, 2, 2)
fitbit.groupby('Work_or_Weekend')['Steps'].median().plot(kind='bar', color='lightgreen')
plt.xlabel('Day Type')
plt.ylabel('Median Steps')
plt.title('Steps on Weekends vs Workdays')
plt.xticks(range(2), ['Weekend', 'Workday'])

# Show plots
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
features = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']
plt.figure(figsize=(20, 7))

for i, feature in enumerate(features, 1):
    plt.subplot(1, 3, i)
    plt.scatter(fitbit[feature], fitbit['Steps'], color='skyblue')
    plt.xlabel(feature)
    plt.ylabel('Steps')
    plt.title(f'Steps vs {feature}')
    plt.grid(True)

plt.show()
```

#### Use the right visulalization to show the distribution of column `Age`.

```python
# your code here
plt.hist(titanic['Age'].dropna(), bins=10, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.grid(True)
plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
plt.figure(figsize=(20, 6))

bins_values = [10, 20, 50]

for i, bins in enumerate(bins_values, 1):
    plt.subplot(1, 3, i)
    plt.hist(titanic['Age'].dropna(), bins=bins, color='skyblue')
    plt.xlabel('Age')
    plt.ylabel('Frequency')
    plt.title(f'Distribution of Age (Bins={bins})')
    plt.grid(True)

plt.show()
```

#### How does the bin size affect your plot?

```python
"""
Smaller bins provide more detailed information about the distribution, especially for datasets with a wide range of values.
Larger bins result in a smoother curve and a more generalized view of the data. They help identify broader patterns in the distribution.
"""
```

#### Use seaborn to show the distribution of column `Age`.

```python
# your code here
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(titanic['Age'].dropna(), color='skyblue', kde=True)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.grid(True)
plt.show()
```

#### Use the right plot to visualize column `Gender`. There are 2 ways of doing it. Do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib
# your code here
plt.bar(titanic['Gender'].value_counts().index, titanic['Gender'].value_counts().values, color=['skyblue', 'lightgreen'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.grid(True)
plt.show()
```

```python
# Method 2 - seaborn
# your code here

sns.countplot(x='Gender', data=titanic, palette=['skyblue', 'lightgreen'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.grid(True)
plt.show()
```

#### Use the right plot to visualize the column `Pclass`.

```python
# your code here
sns.countplot(x='Pclass', data=titanic, palette='Set1')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.title('Pclass Distribution')
plt.grid(True)
plt.show()
```
